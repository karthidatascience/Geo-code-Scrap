```python
# Open the file in read mode
with open('/content/harrisburg.txt', 'r') as file:
    # Read the contents of the file
    contents = file.read()
    
    # Print the contents
# print(contents)
import time
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
dict=[]
s1=re.findall(r'href="\/ho[\d+\/\w+\-\?\=\&\;\%\"]+',contents)
for i in s1:
  time.sleep(2)
  s2=i.replace('href="/','')
  s3=f'https://www.hotels.com/{s2}'
  print(s3)
 
  page1=requests.get(s3)
  soup2=BeautifulSoup(page1.content,'lxml')
  try:
    soup3=str(soup2)
    s1=re.findall(r"(\d+\srooms)",soup3)
    for rooms in s1:
      print(rooms)
  except:
    rooms=''
  hotel_name=soup2.find_all('h1',class_="uitk-heading uitk-heading-3")
  for i in hotel_name:
    name_=i.text
    print(name_)
  location=soup2.find_all('div',class_="uitk-text uitk-type-300 uitk-text-default-theme uitk-layout-flex-item uitk-layout-flex-item-flex-basis-full_width")
  for j in location:
    print(j.text)
    address=j.text.split(',')[0]
    print(address)
    address1=j.text.split(',')[1]
    print(address1)
    city_state=j.text.split(',')[2]
    # print(city_state)
    # city_state_1=j.text.split(',')[3]
    city_state_2=f'{address1},{city_state}'
    print(city_state_2)
    zip_code=j.text.split(',')[-1]
    print(zip_code)

    dictionary = {
          'Hotel Name': name_,
          'Adress': address,
          'City&State': city_state_2,
          'Zip Code': zip_code,
          'Room counts': rooms,
          'links':s3,
          }
    dict.append(dictionary)
    df = pd.DataFrame(dict)
    df.to_csv('D3.csv', index=False)


  
  
```
